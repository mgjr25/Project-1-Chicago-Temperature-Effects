```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("data.cityofchicago.org", "JKjWvr7ChbkqB7fFriD0e9dsZ")


# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.cityofchicago.org,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("8v9j-bter", limit=2000)

# Convert to pandas DataFrame
traffic_segments_df = pd.DataFrame.from_records(results)
traffic_segments_df.to_csv("traffic_segments")
traffic_segments_df
```

```python
results_list = client.get("kkgn-a2j4",where="last_update > '2013-01-01T00:00:00.000' AND last_update < '2013-12-25T23:00:00.000'", limit=2000)

# Convert to pandas DataFrame
historical_df = pd.DataFrame.from_records(results_list)
historical_df
```

```python
merged_df=pd.merge(traffic_segments_df,historical_df,left_on='segmentid',right_on='segment_id')
merged_df
```
